## **2. Literature Review**

### **2.1 Type 2 Diabetes and Early Treatment Failure**

Type 2 diabetes mellitus (T2DM) is a chronic, progressive metabolic disorder characterized by insulin resistance and impaired pancreatic β-cell function (American Diabetes Association, 2023). Although a wide range of pharmacological and lifestyle interventions exist, a considerable proportion of patients fail to achieve or maintain glycaemic targets during the early phase of treatment. This phenomenon, commonly referred to as early treatment failure, is typically defined as the inability to reach clinically recommended HbA1c thresholds within an expected treatment window or the rapid deterioration of glycaemic control after therapy initiation (Davies et al., 2018).

Early treatment failure has been associated with increased risks of microvascular and macrovascular complications, reduced quality of life, and escalating healthcare costs (Holman et al., 2008). Factors contributing to early treatment failure include baseline metabolic heterogeneity, delayed treatment intensification, poor adherence, lifestyle behaviors, and unobserved physiological differences among patients. Traditional clinical monitoring often identifies failure only after substantial deterioration has occurred, underscoring the need for predictive approaches capable of identifying high-risk individuals earlier in the treatment pathway.



### **2.2 Conventional Clinical Trial Designs in Diabetes**

Randomized controlled trials (RCTs) remain the cornerstone of evidence generation in diabetes research. Conventional RCTs typically employ fixed designs with static randomization, predetermined treatment arms, and predefined endpoints, offering strong internal validity and regulatory acceptance (Pocock, 2013). However, such designs primarily focus on average treatment effects, which may obscure clinically meaningful heterogeneity in individual patient responses.

In the context of T2DM, fixed trial designs often lack mechanisms for modifying treatment strategies based on interim patient outcomes. As a result, participants may continue receiving ineffective therapies despite emerging evidence of poor response. This limitation reduces both trial efficiency and ethical responsiveness, particularly in chronic diseases where treatment effects evolve over time. Consequently, there is growing recognition that traditional trial designs are poorly suited for evaluating personalized and adaptive treatment strategies in diabetes care. ([https://pubmed.ncbi.nlm.nih.gov/41264554/])

These findings highlight the limitations of traditional designs for **dynamic treatment optimization** and motivate the inclusion of adaptive strategies that can better personalize interventions in chronic diseases like T2DM.



### **2.3 Adaptive Clinical Trials in Healthcare**

Adaptive clinical trials introduce flexibility by allowing pre-specified modifications to trial conduct based on accumulating data, while maintaining statistical validity (Chow & Chang, 2011). These modifications may include adaptive randomization, sample size re-estimation, early stopping, or treatment arm modification. Adaptive designs have demonstrated advantages in efficiency, ethical treatment allocation, and accelerated decision-making across several therapeutic areas.

In healthcare research, adaptive trials are particularly valuable in conditions characterized by patient heterogeneity and dynamic treatment responses, such as T2DM. Despite their suitability, adaptive designs remain underutilized in diabetes research, with most applications limited to dose-finding or interim monitoring rather than full decision-adaptive frameworks (Kairalla et al., 2012). Barriers to adoption include methodological complexity, regulatory challenges, and the need for advanced analytical tools capable of supporting real-time adaptation.



### **2.4 Artificial Intelligence in Clinical Decision Support**

Artificial intelligence (AI), particularly machine learning, has emerged as a transformative approach in clinical decision support systems (CDSS). AI algorithms are capable of processing high-dimensional clinical data to identify complex, nonlinear patterns associated with disease progression and treatment outcomes (Esteva et al., 2019). In diabetes management, AI applications have been used for risk prediction, insulin dose optimization, and personalized treatment recommendations. ([https://pubmed.ncbi.nlm.nih.gov/30441340/])

AI-driven CDSS enable proactive clinical decision-making by predicting adverse outcomes before they manifest clinically. Studies have demonstrated that machine learning models can outperform traditional statistical approaches in predicting glycaemic deterioration and treatment failure (Kavakiotis et al., 2017). However, most AI models are developed and validated retrospectively and are rarely integrated directly into clinical trial frameworks to guide adaptive treatment decisions in real time. ([https://www.sciencedirect.com/science/article/pii/S1386505624004465?utm_source=chatgpt.com])

Additionally, meta-analyses have shown that AI-based closed-loop systems, which integrate continuous glucose monitors and automated insulin delivery technologies, improve clinical outcomes such as time in target glucose range compared with traditional care. ([https://link.springer.com/article/10.1186/s13098-025-01819-0?utm_source=chatgpt.com])

Despite these advances, challenges remain. Issues related to data quality, model bias, and the need for robust external validation limit the broad clinical implementation of AI systems. ([https://www.nature.com/articles/s41746-024-01034-7?utm_source=chatgpt.com]) Nevertheless, the potential for AI to inform individualized treatment responses is increasingly recognized as transformative in chronic disease management.

# Feature Engineering
## Motivation

Raw clinical variables often fail to capture latent risk patterns relevant to early treatment failure. Feature engineering was therefore applied to enhance model sensitivity and clinical interpretability.

## Derived Features

**Several derived variables were created:**

* Adherence risk: Defined as one minus the medication adherence rate, emphasizing non-adherence as a risk signal.

* Normalized glucose variability: Standardization was applied to facilitate model convergence and comparability across features.

* Activity per 1,000 steps: Rescaling step counts improves numerical stability and interpretability.

* Age-centered variable: Centering reduces collinearity and improves model calibration.

These transformations mimic standard preprocessing steps used in clinical machine learning pipelines.

## Feature Selection

The final feature set included demographic, metabolic, behavioral, and physiological variables, reflecting a multimodal digital biomarker approach. Missing values were imputed using median values, a robust strategy commonly employed in clinical datasets.

## Feature engineering for ML (7-day variability, HRV mean etc.) codes

```python
#Simulate longitudinal CGM time series (7 days, 96 readings/day)

n_days = 7
points_per_day = 96

cgm_series = {}

for _, row in clinical_data.iterrows():
    base_glucose = row['avg_daily_glucose']
    variability = row['glucose_variability']

    noise = np.random.normal(0, variability, n_days * points_per_day)
    trend = np.linspace(0, np.random.normal(5, 2), n_days * points_per_day)

    series = np.clip(base_glucose + noise + trend, 60, 400)
    cgm_series[row['participant_id']] = series
```

```python
# Feature engineering with longitudinal CGM features

clinical_data_feat = clinical_data.copy()

clinical_data_feat['adherence_risk'] = 1 - clinical_data_feat['medication_adherence_rate']

clinical_data_feat['glucose_var_norm'] = (
    clinical_data_feat['glucose_variability'] -
    clinical_data_feat['glucose_variability'].mean()
) / clinical_data_feat['glucose_variability'].std()

clinical_data_feat['activity_per_1000'] = clinical_data_feat['avg_daily_steps'] / 1000.0


clinical_data_feat['hba1c_x_nonadherence'] = (
    clinical_data_feat['baseline_HbA1c'] *
    clinical_data_feat['adherence_risk']
)

clinical_data_feat['glucvar_x_nonadherence'] = (
    clinical_data_feat['glucose_variability'] *
    clinical_data_feat['adherence_risk']
)

clinical_data_feat['cgm_mean_7d'] = [
    cgm_series[pid].mean() for pid in clinical_data_feat['participant_id']
]

clinical_data_feat['cgm_sd_7d'] = [
    cgm_series[pid].std() for pid in clinical_data_feat['participant_id']
]

clinical_data_feat['cgm_cv_7d'] = (
    clinical_data_feat['cgm_sd_7d'] /
    (clinical_data_feat['cgm_mean_7d'] + 1e-3)
)

features = [
    'baseline_HbA1c',
    'avg_daily_glucose',
    'glucose_variability',
    'glucose_var_norm',
    'medication_adherence_rate',
    'adherence_risk',
    'hba1c_x_nonadherence',
    'glucvar_x_nonadherence',
    'cgm_mean_7d',
    'cgm_sd_7d',
    'cgm_cv_7d',
    'activity_per_1000',
    'heart_rate_variability'
]

X = clinical_data_feat[features].fillna(
    clinical_data_feat[features].median()
)

y = clinical_data_feat['early_treatment_failure']
```

```python
#Simulate time-to-treatment failure

baseline_hazard = 0.02

time_to_failure = np.random.exponential(
    scale=1 / (baseline_hazard * (clinical_data['prob_failure_sim'] + 1e-3))
)

clinical_data_feat['time_to_failure'] = np.round(time_to_failure, 2)
clinical_data_feat['event'] = clinical_data_feat['early_treatment_failure']
```

## Predictive Modeling and Model Selection
## Modeling Objective

The objective of this modeling stage is to predict early treatment failure in patients with Type 2 diabetes using baseline clinical characteristics, wearable-derived biomarkers, and behavioral indicators. Given the binary nature of the outcome, supervised classification models are employed and evaluated based on their discriminative performance.

## Models Evaluated

Three complementary machine learning models are trained and compared to capture both linear and non-linear relationships in the data:

### Logistic Regression

Logistic regression is selected as a baseline, interpretable model commonly used in clinical research. It assumes a linear relationship between predictors and the log-odds of the outcome, making it suitable for benchmarking and coefficient-based interpretation.

Before training, input features are standardized to ensure numerical stability and appropriate convergence. The model is trained with an increased maximum iteration limit to guarantee convergence given the number of predictors.

### Random Forest Classifier

A random forest classifier is used to model non-linear interactions and complex feature relationships without requiring feature scaling. By aggregating predictions from 200 decision trees, the model reduces variance and is robust to noise commonly observed in wearable and behavioral data.

Random forests are particularly well-suited to heterogeneous clinical data and often perform well when interactions between physiological and behavioral variables exist.

### Extreme Gradient Boosting (XGBoost)

XGBoost is included as a state-of-the-art ensemble learning algorithm capable of capturing subtle non-linear patterns and higher-order interactions. Its regularization mechanisms help prevent overfitting, making it a strong candidate for clinical prediction tasks.

The model is trained using a log-loss objective function, consistent with probabilistic binary classification, and configured to efficiently utilize available computational resources.

## Model Evaluation Strategy

Model performance is assessed using the Area Under the Receiver Operating Characteristic Curve (AUC-ROC). AUC is a threshold-independent metric widely used in clinical prediction modeling, reflecting a model’s ability to correctly rank patients who experience treatment failure versus those who do not.

The test set is held out during training and preserved the original class distribution via stratified sampling to ensure unbiased performance estimation.

```python

# Initialize models
models = {
    'LogisticRegression': LogisticRegression(random_state=RANDOM_SEED, max_iter=1000),
    'RandomForest': RandomForestClassifier(n_estimators=200, random_state=RANDOM_SEED, n_jobs=-1),
    'XGBoost': xgb.XGBClassifier(
        n_estimators=500,
        max_depth=3,
        learning_rate=0.05,
        subsample=0.8,
        colsample_bytree=0.8,
        min_child_weight=5,
        gamma=0.1,
        reg_alpha=0.5,
        reg_lambda=1.0,
        scale_pos_weight=(1 - y_train.mean()) / y_train.mean(),
        eval_metric='auc',
        random_state=RANDOM_SEED,
        n_jobs=-1
    )
}

# Train, predict, and compute AUC for each model
model_aucs = {}

for name, model in models.items():
    # Fit model
    if name == 'LogisticRegression':
        model.fit(X_train_scaled, y_train)  # Logistic Regression uses scaled features
        probs = model.predict_proba(X_test_scaled)[:, 1]
    else:
        model.fit(X_train, y_train)  # Tree-based models can use raw features
        probs = model.predict_proba(X_test)[:, 1]
    
    # Compute AUC
    auc = roc_auc_score(y_test, probs)
    model_aucs[name] = auc
    print(f"{name} AUC: {auc:.3f}")

# Automatically select the best model
best_model_name = max(model_aucs, key=model_aucs.get)
best_auc = model_aucs[best_model_name]

print("\nSelected best model:", best_model_name, "with AUC:", round(best_auc, 3))


```

## 3.0. **Objective 3: Assess the Impact of Real-Time AI-Assisted Treatment Adjustments**

### 3.1 **Overview**

This objective evaluated the effectiveness of **real-time AI-assisted treatment adjustments** in reducing early treatment failure compared with standard care. The analysis simulated adaptive clinical decision-making within an AI-Adaptive trial arm, using individualized risk scores to guide timely interventions.


### 3.2 **Baseline Outcome Preservation**

To enable a valid comparison, the original early treatment failure outcome was preserved as a reference variable. This ensured that any observed changes in failure rates could be attributed solely to the simulated AI-driven intervention rather than alterations in baseline outcome definitions.




### 3.4  **AI-Assisted Intervention Logic**

#### 3.4.1  **Risk-Stratified Treatment Adjustment**

The AI intervention was applied exclusively to participants assigned to the **AI-Adaptive trial arm** and operated under a predefined, clinically interpretable risk-stratification framework:

* **High-risk participants (risk score ≥ 0.70):**
  Early treatment failure probability was reduced by **40%**, reflecting aggressive real-time treatment optimization.

* **Moderate-risk participants (risk score 0.40–0.69):**
  Failure probability was reduced by **20%**, simulating targeted clinical review and early corrective action.

* **Low-risk participants (< 0.40) and Control arm:**
  No AI intervention was applied, and outcomes followed standard care trajectories.

This rule-based approach mirrors real-world clinical decision support systems where intervention intensity scales with patient risk.


```python
# Copy original outcome for comparison
clinical_data['treatment_failure_original'] = clinical_data_feat['early_treatment_failure']

# Apply AI intervention
def apply_ai_effect(row):
    if row['trial_arm'] == 'AI-Adaptive':
        # If risk score >= 0.7 (high risk), reduce failure probability by 40%
        if row['failure_risk_score'] >= 0.7:
            # Randomly determine if failure is prevented
            return 0 if np.random.rand() < 0.4 else row['early_treatment_failure']
        # Moderate risk: risk score 0.4–0.7, 20% reduction
        elif 0.4 <= row['failure_risk_score'] < 0.7:
            return 0 if np.random.rand() < 0.2 else row['early_treatment_failure']
    return row['early_treatment_failure']  # Control arm or low risk

clinical_data['treatment_failure_ai'] = clinical_data.apply(apply_ai_effect, axis=1)
```
